Plot the variation of prices fort all observations in dataset

```python
PriceList = brentoil['Price'].tolist()
DateList  = brentoil['Date'].tolist()

plt.plot(PriceList, color='red', label = 'Variation of Prices from 1987 to 2019')

plt.xlabel('(n-1) observations')
plt.ylabel('Change in Price')
plt.title('Change in Prices')
plt.legend()
plt.show()
```

Plot the variation in prices over the years

```python
#convert Date to datetime object
brentoil.Date = pd.to_datetime(brentoil.Date)

#Extract year from datetime object Date
brentoil['year']=pd.DatetimeIndex(brentoil.Date).year
brentoil['month']=pd.DatetimeIndex(brentoil.Date).month

res=brentoil[['Price','year']].groupby(['year']).sum()

plt.plot(res, label = 'Variation of Prices from 1987 to 2019')

plt.xlabel('years')
plt.ylabel('Change in Price')
plt.title('Change in Prices')
plt.legend()
plt.show()
```

Prices have fallen sharply after 2007, plot the yearly variation of prices between 2007 and 2019 

```python
res1 = res.loc[2007:2019]

plt.plot(res1)
plt.show()
```

Plot the variations of Prices over months

```python
res2=brentoil[["Price","month"]].groupby(["month"]).mean()

plt.plot(res2)
plt.xlabel('months')
plt.ylabel('changes in prices')
plt.show()
```
