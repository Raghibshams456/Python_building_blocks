```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pulsars = pd.read_csv(r'pulsar_stars.csv')
```

```python
pulsars.info()
```

```python
pulsars.describe()
```

```python
pulsars.corr()
```

```python
#Using pairplot

sns.pairplot(data=pulsars, palette = "husl", hue="target_class",
            vars=[" Mean of the integrated profile",
                   " Excess kurtosis of the integrated profile",
                   " Skewness of the integrated profile",
                   " Mean of the DM-SNR curve",
                   " Excess kurtosis of the DM-SNR curve",
                   " Skewness of the DM-SNR curve"])


plt.suptitle("PairPlot of Data Without Std. Dev. Fields",fontsize=18)
plt.show() 
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(data=pulsars.corr(),annot=True,cmap="bone",linewidths=1,fmt=".2f",linecolor="gray")
plt.title("Correlation Map",fontsize=20)
plt.tight_layout()
plt.show()      # lightest and darkest cells are most correlated ones
```

```python
plt.figure(figsize=(16,10))

plt.subplot(2,2,1)
sns.violinplot(data=pulsars,y=" Mean of the integrated profile",x="target_class")

plt.subplot(2,2,2)
sns.violinplot(data=pulsars,y=" Mean of the DM-SNR curve",x="target_class")

plt.subplot(2,2,3)
sns.violinplot(data=pulsars,y=" Standard deviation of the integrated profile",x="target_class")

plt.subplot(2,2,4)
sns.violinplot(data=pulsars,y=" Standard deviation of the DM-SNR curve",x="target_class")


plt.suptitle("ViolinPlot",fontsize=20)

plt.show()
```
