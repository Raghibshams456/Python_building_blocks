Find out how many crossings at each border:

```python
df.Border.value_counts()
```

Now Find out the measure type at US-Canada Border

```python
res=df[["Border","Measure","Value"]]
res[res.Border=='US-Canada Border'].groupby(['Measure']).sum().unstack()
```

Now Find out the measure type at US-Mexico Border

```python
res[res.Border=='US-Mexico Border'].groupby(['Measure']).sum().unstack()
```

Find out the crossing at each Port Name

```python
res1=df[["Port Name","Measure","Value"]]
res2=res1.groupby(['Port Name']).sum()
res2
```

Now find out the maximum crossing Port by all measures

```python
res2.sort_values(by='Value', ascending=False).head(1)
```

Find out how many trucks crossed Calexico East

```python
res3=df[["Port Name","Measure","Value"]]
res4=res3[res3['Port Name']=='Calexico East']
res4[res4['Measure']=='Trucks'].groupby('Measure').sum()
```
