Discover what is the mean number of area lost to forest fire per state

```python
df.groupby('state').number.mean()
```

Discover what is the mean number of area lost to forest fire per month

```python
df.groupby('month').number.mean().sort_values()
```

Discover what is the mean number of area lost to forest fire per year

```python
df.groupby('year').number.mean().sort_values()
```

Discover which state area was lost highest to the forest fire groupby year

```python
res =df[["year","state","month","number"]].groupby(['year','state']).sum().unstack()
res
```

```python
#Now to answer the above question

max(res.number)
```

```python
#To find out how much it has lost and in which year,
res.number[["Tocantins"]].sort_values(by='Tocantins',ascending=False).head(1)
```

```python
#Likewise least area lost to fires by state, by year is:
min(res.number)
```
