sort countries with highest systemic_crisis

```python
#view only countries with highest systemic_crisis

sc = df[["country","systemic_crisis"]]
sc
```

```python
sc.groupby(["country"]).sum().sort_values(by='systemic_crisis', ascending=False)
```

sort countries with lowest sovereign_external_debt_default

```python
df[["country","sovereign_external_debt_default"]].groupby(["country"]).sum().sort_values(by='sovereign_external_debt_default')
```

Top 5 countries with highest currency_crises and inflation_crises

```python
df[["country", "currency_crises","inflation_crises"]].groupby(["country"]).sum().sort_values(by ='currency_crises', ascending=False).head(5)
```

```python
res = df[["country", "inflation_annual_cpi"]]
res=res.groupby(["country"])
round(res.mean())
```
