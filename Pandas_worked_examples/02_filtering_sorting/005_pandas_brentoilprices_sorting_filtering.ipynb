```python
#convert Date object to datetime object

df.Date=pd.to_datetime(df.Date)
df.dtypes
```

```python
#extracting year and month seperately

df['year']=pd.DatetimeIndex(df.Date).year
df['month']=pd.DatetimeIndex(df.Date).month
```

```python
df.head()
```

```python
#find out year by year variation of brent oil prices
df[['Price','year']].groupby(['year']).sum()
```

```python
#find out what month was the mean price highest

df[["Price","month"]].groupby(["month"]).mean().sort_values(by='Price', ascending=False)
```

```python
df.describe()
```

```python
#Variation of price from previous month

res=df[["Price","Date"]].set_index('Date').diff()
res=res.dropna()
res
```

```python
#highest price increase happened on following month and Date
res.loc[res.Price.idxmax()]
```

```python
#highest price decrease in brent oil happened on following date
res.loc[res.Price.idxmin()]
```
