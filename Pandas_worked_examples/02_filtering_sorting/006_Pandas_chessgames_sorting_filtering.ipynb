```python
chessdf = df.copy()
```

```python
#dropping columns with nan values more than 50%
chessdf.dropna(thresh=0.5, inplace=True)
```

```python
chessdf
```

#dropping unrequired columns like id, rated, created_at, last_move_at,increment_code, white_id,black_id, opening_eco,opening_ply


```python
chessdf.drop(columns=['id', 'rated', 'created_at', 'last_move_at','increment_code', 'white_id','black_id', 'opening_eco','opening_ply'], inplace=True)
```

```python
chessdf.head()
```

```python
#Count number of wins for whites and blacks 

chessdf.winner.value_counts()
```

```python
#Findout number of wins with starting move being e4
chessdf.moves.str.startswith('e4').value_counts('True')
```

```python
#How many times has the game won by mate by black

res=chessdf[["victory_status","winner"]].groupby(['victory_status', 'winner']).size()
res
```

Conclusion: Game was won by mate by black 2981, drawn 906 times and resigned 5303 times

```python
#what is the least number of turns in which the game was won

res1 = chessdf[["turns","victory_status","winner"]]
res2=res1.sort_values(by='turns')
res2[res2.victory_status=='mate'].head(20)
```
